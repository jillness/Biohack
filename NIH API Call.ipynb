```python
#load dependencies
import requests
import json
import pandas as pd
import time
```

```python
# URL for API call
#url = "https://api.federalreporter.nih.gov/v1/Projects/search?query=text:cancer biology"
```

```python
# we will need to add all our search terms to this URL
url = "https://api.federalreporter.nih.gov/v1/projects/search?query=text:cancer biology$textFields:terms$textOperator:OR$biotech"
```

```python
# check to see if response is valid
# wait 
print(requests.get(url))
```

```python
response = requests.get(url).json()

# print results from API call
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
response.keys()
```

```python
key_words = []
project_abstract = []
project_title = []
project_number = []
project_start_date = []
project_end_date = []
department = []
agency = []
PI = []
latitude = []
longitude = []
org_name = []
org_city = []
org_state = []
org_zip = []
org_country = []
budget_start_date = []
budget_end_date = []
CFDA_code = []
fiscal_year = []
total_cost = []
i = 0
# range needs to be updated.. I just used 2 for testing purposes so the code would execute faster
for i in range(2):
    response = requests.get(url).json()
    for item in response['items']:
        key_words.append(item['terms'])
        project_abstract.append(item['abstract'])
        project_title.append(item['title'])
        project_number.append(item['projectNumber'])
        project_start_date.append(item['projectStartDate'])
        project_end_date.append(item['projectEndDate'])
        department.append(item['department'])                        
        agency.append(item['agency'])                
        PI.append(item['contactPi'])
        latitude.append(item['latitude'])
        longitude.append(item['longitude'])
        org_name.append(item['orgName'])                 
        org_city.append(item['orgCity'])
        org_state.append(item['orgState'])                
        org_zip.append(item['orgZipCode'])
        org_country.append(item['orgCountry'])               
        budget_start_date.append(item['projectStartDate'])               
        budget_end_date.append(item['projectEndDate'])
        CFDA_code.append(item['cfdaCode'])                      
        fiscal_year.append(item['fy'])                       
        total_cost.append(item['totalCostAmount']) 
i += 1
time.sleep(0.5)
        
print(key_words)
print(PI)
```

```python
# create a dictionary and store data in pandas dataframe
NIH_dict = {
    "Key Words": key_words,
    "Project Abstract": project_abstract,
    "Project Title": project_title,
    "Project Number": project_number,
    "Project Start Date": project_start_date,
    "Project End Date": project_end_date,
    "Department": department,
    "Agency": agency,
    "Principal Investigator": PI,
    "Latitude": latitude,
    "Longitude":longitude,
    "Organization Name": org_name,
    "Organization City": org_city,
    "Organization State": org_state,
    "Organization ZIP": org_zip,
    "Organization Country": org_country,
    "Budget Start Date": budget_start_date,
    "Budget End Date": budget_end_date,
    "CFDA code": CFDA_code,
    "Fiscal Year": fiscal_year,
    "Total Cost": total_cost
}
NIH_data = pd.DataFrame(NIH_dict)
NIH_data.head()
```

```python
# save DF to a csv
NIH_data.to_csv("NIH_data.csv", encoding="utf-8", index=False)
```
