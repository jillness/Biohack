```python
#load dependencies
import requests
import json
import pandas as pd
import time
from math import ceil
```

```python
# key term list
key_terms_df = pd.read_csv("client_keyword_results.csv")
key_terms = key_terms_df['keyterm']
key_terms_df.head()
```

```python
url_list = []

for word in key_terms:
    word = word.strip().replace(' ', '+')
    url = f"https://api.federalreporter.nih.gov/v1/projects/search?query=text:{word}"
    url_list.append(url)

url_list[:5]
```

```python
i = 0 
j = 0
try:
    param = {"offset": i}  #offsets the page number of search results by amount i (start at page i of [total pages])
    response = requests.get('https://api.federalreporter.nih.gov/v1/projects/search?query=text:next+generation+sequencing', params=param).json()  #following requests with page offset

    for item in response['items']:
        print(j, item['terms'])
        j+=1
        
    i += 50
    time.sleep(0.4)
except:
    pass
```

```python
response['items'][0]['abstract']
```

```python
key_terms_df['url'] = url_list
key_terms_df.head()
```

```python
# NIH query using key words from list
# collect the following info 

key_words = []
project_abstract = []
project_title = []
project_number = []
project_start_date = []
project_end_date = []
department = []
agency = []
PI = []
latitude = []
longitude = []
org_name = []
org_city = []
org_state = []
org_zip = []
org_country = []
budget_start_date = []
budget_end_date = []
CFDA_code = []
fiscal_year = []
total_cost = []
seed_term = []

for url in url_list:
    i = 0 
    param = {"offset": i}  #offsets the page number of search results by amount i (start at page i of [total pages])
    response = requests.get(url, params=param).json()  #following requests with page offset
    
    try:
        for item in response['items']:
            seed_term.append(url[66:])
            key_words.append(item['terms'])
            project_abstract.append(item['abstract'])
            project_title.append(item['title'])
            project_number.append(item['projectNumber'])
            project_start_date.append(item['projectStartDate'])
            project_end_date.append(item['projectEndDate'])
            department.append(item['department'])                        
            agency.append(item['agency'])                
            PI.append(item['contactPi'])
            latitude.append(item['latitude'])
            longitude.append(item['longitude'])
            org_name.append(item['orgName'])                 
            org_city.append(item['orgCity'])
            org_state.append(item['orgState'])                
            org_zip.append(item['orgZipCode'])
            org_country.append(item['orgCountry'])               
            budget_start_date.append(item['projectStartDate'])               
            budget_end_date.append(item['projectEndDate'])
            CFDA_code.append(item['cfdaCode'])                      
            fiscal_year.append(item['fy'])                       
            total_cost.append(item['totalCostAmount'])

        i += 50
        time.sleep(0.4)
        
    except:
        print(f"No search results for {url}")
        pass

```

```python
len(key_words)
```

```python
# store in DataFrame
NIH_dict = {
    "Seed Term": seed_term,
    "Key Words": key_words,
    "Project Abstract": project_abstract,
    "Project Title": project_title,
    "Project Number": project_number,
    "Project Start Date": project_start_date,
    "Project End Date": project_end_date,
    "Department": department,
    "Agency": agency,
    "Principal Investigator": PI,
    "Latitude": latitude,
    "Longitude":longitude,
    "Organization Name": org_name,
    "Organization City": org_city,
    "Organization State": org_state,
    "Organization ZIP": org_zip,
    "Organization Country": org_country,
    "Budget Start Date": budget_start_date,
    "Budget End Date": budget_end_date,
    "CFDA code": CFDA_code,
    "Fiscal Year": fiscal_year,
    "Total Cost": total_cost
}
NIH_data = pd.DataFrame(NIH_dict)
NIH_data
```

```python
# save DF to a csv
NIH_data.to_csv("FedReporter.csv", encoding="utf-8", index=False)
```
